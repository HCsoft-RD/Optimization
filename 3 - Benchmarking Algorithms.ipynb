# Benchmarking optimization algorithms in Python

Now it is time to see how different algorithms perform on a given set of funtions. In order to compare them we have build the OptBenchmark class, a Dashboard that will help us manage the data collection process. 


```python
benchmark = OptBenchmark()
benchmark.widget
```

<img src="benchmark_data/widget.png">

As you can see, this class is meant to be used as an interactive dashboard. It has many options about how to perform the benchmark.

## 1. The main menu

In the following image, we have the following widgets located at the left of the Dashboard.

<img src="benchmark_data/status.png">

- **Version**: Textbox when you can write a name for the algorithm currently tested. This allows to distinguis between two different set of paramters for the same algorithm. 
- **Run button**: Starts the benchmark.
- **Progress bars**: Show the current state of the benchmarking calculations. 
- **Max time**: Sets the mast time a single function optimization can last. It is only taken into account is Time is an active end condition. 
- **Rounds**: How many rounds the algorithm will be tested against every objective function.

## 2. End condition options

These widgets allows us to set when we want the optimization process to finish. We can set up to 4 complementary conditions for ending an optimization process. Clicking on a toggle button will activate one of the following options:

- **Find best**: The optimization process will stop once the minimum value of the objective function has been found. 
- **Max epochs**: Maximum number of iterations an algorithm can perform before finishing the optimization process.
- **Time**: Maximum time to be spend when trying to optimize an objective function.
- **Samples**: Maximum number of function evaluations allowed.

<img src="benchmark_data/end_codition.png">

# 3. Found options

This menu is intended for specifiying how a successful optimization will be defined. The widgets in this submenu allow us to specify the sensibility for triggering the "Find best" condition, and therefore for considering an objective function solved.

There are two complementary ways of defining an and condition:

- **Distance**: The euclidean distance between the position of the best solution found and the position of the global minimum is less than a threshold.
- **Error**: The difference between the best value found and the global minimum is less than a threshold.

This threshold n is set using the slider widget. If the log10 error checkbox is clicked the threshold value will be 10^n. If the checkbox is not active the trheshold value will be n.

<img src="benchmark_data/found.png">

# 4. Save options

It is possible to automatically save the results of a given benchmarking process as it is being performed using the save options widgets.

It is also possible to specify the name of the file where the data will be stored. This file will be created inside a data/ folder that must be located in the same directory directory as the notebook.

The togglebuttons menu is used for selecting how often the results of the benchmark will be saved. 

- **Round**: Save the results every time that all the algorithms have finished all the objective functions once. 
- **Function**: Save after every algorithm finishes one objective function. 
- **Algorithm**: Save when an algorithm finishes all the objective functions.

The slider labeled "Every" selects how many times the condition selected on the togglemenu has to be met before saving the results.

<img src="benchmark_data/save.png">

```python
#data wrangling imports
from os import listdir
from os.path import isfile, join
import time
import pandas as pd
import numpy as np
from IPython.core.display import clear_output
from shaolin.core.dashboard import Dashboard
from shaolin.dashboards.colormap import ColormapPicker
from test_functions import TestFunctions
from optimizer import Optimizer
```

```python
class OptBenchmark(Dashboard):
    
    def __init__(self,
                 time=3000,
                 rounds=100,
                 min_error=-6.,
                 min_dist=-3.,
                 width="145em",
                 height="85em",
                 sleep=0.1,
                 **kwargs):


        self._trans = {'differential_evolution':'de',
                       'basin_hopping':'bh',
                      }
        
        test_funcs = TestFunctions(name='test_funcs',mode='interactive')
        optimizer = Optimizer(test_funcs.functions['eggholder'],mode='interactive',name="optimizer_d",**kwargs)
        self._target_name = optimizer.name
        self._height = height
        self._width = width
        self._filtered_df  = None
        dashboard = ['c$N=mpl_plot',
                     [optimizer,
                      test_funcs,
                      ["r$N=bench_row",[["c$n=bench_col_1",[["r$N=run_row",["btn$d=Run&N=run_button","text$d=Version&val="]],
                                         "ip$d=Run&n=run_pb&min=0&max="+str(int(rounds))+"&val=0",
                                         "ip$d=Function&n=func_pb&min=0&max="+str(len(test_funcs.functions.keys()))+"&val=0",
                                         
                                         "(0.,3600,1,"+str(int(time))+")$d=Max time",
                                         "(1,1000,1,"+str(int(rounds))+")$d=Rounds"]],
                                        ["c$n=end_col",["###End condition$n=end_title",
                                                        "[True]$d=Find best&n=best_tog",
                                                        "[False]$d=Max epochs&n=epoch_tog",
                                                        "[False]$d=Time&n=time_tog",
                                                        "[True]$d=Samples&n=sample_tog"]]
                                        ,["c$N=found_col",["###Found condition$n=found_title",
                                                           ["r$n=mini_row",["[False]$d=Distance&n=dist_tog",
                                                            "[True]$d=Value&n=val_tog"]],
                                                            "(-20.,100.,0.1,"+str(float(min_error))+")$d=Min error",
                                                            "(-20.,100.,0.1,"+str(float(min_dist))+")$d=Min distance",
                                                            "True$d=Log10 error",
                                                           ]
                                          ]
                                           
                       ,["r$N=filters_row",[["c$n=mem_col",["text$d=File name&n=save_name",
                                                            "togs$d=Save&n=val_every&o=['Round','Function','Algorithm']",
                                                            "(1,250,1,1)$d=Every&n=num_every"]],
                                                          "selmul$d=Algorithms&N=algos_sel&o="+str(optimizer.algos)]]]],             
                      'togs$n=control_togs&o=["Benchmark","Optimizer","Test functions"]','html$N=plot'
                      
                     ]
                    ]
        Dashboard.__init__(self, dashboard, mode='interactive', **kwargs)
        #default benchmark functions
        self.set_default_benchmark()
        #self.show_filters.observe(self._toggle_filters)
        self._html_display_value = ''
        #layout fixes
        self.plot.widget.layout.width = "100%"
        self.plot.target.layout.width = "100%"
        self.version.widget.layout.width = "12em"
        self.version.target.layout.width = "12em"
      
        self.control_togs.observe(self.update_layout)
        self.dist_tog.observe(self._update_error_layout)
        self.val_tog.observe(self._update_error_layout)
        self.run_button.observe(self.run_benchmark)
        self.optimizer.run_btn.visible = False
        self._update_error_layout()
        self.update_layout()
        self.reset_data()
        self.run_details = {}

    def set_default_benchmark(self):
        """Defaults settings for the test functions"""
        vals = list(self.test_funcs.wiki.func_sel.target.options)

        vals.pop(vals.index('bukin6'))
        self.test_funcs.wiki.func_sel.value = tuple(vals)
        self.test_funcs.multi.ras_tog.value = False
        self.test_funcs.multi.dv2_tog.value = False
        self.test_funcs.multi.ras_dim.value = (30, 61)
        self.test_funcs.multi.ras_step.value = 30
        self.test_funcs.multi.ros_tog.value = False
        self.test_funcs.multi.ros_dim.value = (30, 61)
        self.test_funcs.multi.ros_step.value = 30
        self.test_funcs.multi.update()
        self.test_funcs.update()
        
    
    def reset_pbar(self):
        """Reset progress bar"""
        if self.rounds.value < self.run_pb.value:
            self.run_pb.value = self.rounds.value 
        self.run_pb.target.max = self.rounds.value 
        self.run_pb.value = 1
        if len(self.funcs.keys()) < self.func_pb.value:
            self.func_pb.value = len(self.funcs.keys())
        self.run_pb.target.max = len(self.funcs.keys())
        self.run_pb.value = 1
        
    
            
    def _update_error_layout(self,_=None):
        """Visual updating of the layout"""
        if self.dist_tog.value:
            self.min_distance.visible = True
        else:
            self.min_distance.visible = False
        if self.val_tog.value:
            self.min_error.visible = True
        else:
            self.min_error.visible = False
    
    def compute_error(self):
        """Compute error values for the benchmark functions"""
        fo_val = False
        fo_dist = False
        error = np.nan
        dist = np.nan
        error = self.optimizer.best[1]-self.optimizer.pot_func.benchmark[1]
         
        if self.val_tog.value:
            if self.log10_error.value:
                fo_val = error<10**self.min_error.value
            else:
                fo_val = error<self.min_error.value
        if not self.optimizer.best[0] is None:
            dist = np.linalg.norm(self.optimizer.pot_func.benchmark[0]-self.optimizer.best[0])
        else:
            dist = 1e10 
        dist = max(dist,1e-10)
        if self.dist_tog.value:
            if self.log10_error.value:
                fo_dist = dist<10**self.min_distance.value
            else:
                fo_dist = dist<self.min_distance.value        
        found = fo_val or fo_dist
        assert(not np.isnan(error))
        return error,dist,found
    
    
    def end_condition(self):
        """Callback that returns true when the finishing conditions are met"""
        end = False
        if self.best_tog.value:
            end = self.compute_error()[2]#best found
        #max epoch reached
        if self.epoch_tog.value:
            end = end or self.optimizer.epoch >= self.optimizer.max_epoch.value
        #max reads reached
        if self.sample_tog.value:
            end = end or self.optimizer.pot_func.n_reads > self.optimizer.max_reads.value
        #max time exceded
        if self.time_tog.value:
            end = end or time.time()-self._run_start > self.max_time.value
        return end
    
    def reset_data(self):
        """Restarts the data collection process"""
        funcs = list(self.funcs.keys())
        ids = []
        for suf in self._trans.values():
            for f in funcs:
                n = f+'_'+suf
                ids.append(n+'_'+self.version.value if self.version.value !='' else n)
        self._id_vals = ids
        runs = np.arange(self.rounds.value).tolist()
        res_ix = pd.MultiIndex.from_product([ids,runs])
        res_ix.levels[0].name = 'id_val'
        res_ix.levels[1].name = 'run'
        self.results = pd.DataFrame(index=res_ix,columns=['time','reads','value','x','found','dist','error','function','algo','id','epoch'])
        self.results = self.results.reset_index()
        self.results = self.results.fillna(0)
        self.results.index = res_ix

    def update_layout(self,_=None):
        """Updating of GUI"""
        if self.control_togs.value == 'Optimizer':
            self.test_funcs.visible = False
            self.optimizer_d.visible = True
            self.bench_row.visible = False
        elif self.control_togs.value == 'Test functions':
            self.test_funcs.visible = True
            self.optimizer_d.visible = False
            self.bench_row.visible = False
        elif self.control_togs.value == 'Benchmark':
            self.test_funcs.visible = False
            self.optimizer_d.visible = False
            self.bench_row.visible = True
    
    @property
    def funcs(self):
        """Selected test functions"""
        return self.test_funcs.functions
    
    @property
    def optimizer(self):
        """Selected optimizer dashboard"""
        return getattr(self,self._target_name)
    
    
    def run_benchmark(self,_=None):
        """Test the selected solvers"""
        self.reset_data()
        self.reset_pbar()
        start= time.time()
        for k,(algo,suf) in enumerate(self._trans.items()):
            if algo in self.algos_sel.value:
                self.func_pb.value = 1
                for i in range(self.rounds.value):                   
                    for j,(_name,func) in enumerate(self.funcs.items()):
                        name = _name+'_'+suf
                        full_name = name if self.version.value==''else name+'_'+self.version.value
                        version = suf if self.version.value==''else suf+'_'+self.version.value
                        
                        clear_output(True)
                        print("Calculating {} round {}".format(_name,i))
                        x,val,reads,elapsed_time = self.run_one(name=algo,func=func)
                        self.results.loc[(full_name,i),"epoch"] = self.optimizer.epoch
                        self.results.loc[(full_name,i),"version"] = str(version)
                        #full_id is func_version
                        self.results.loc[(full_name,i),"full_id"] = str(full_name)
                        self.results.loc[(full_name,i),"function"] = str(_name)
                        self.results.loc[(full_name,i),"algo"] = str(algo)
                        self.results.loc[(full_name,i),"x"] = str(x.tolist())
                        #noise for avoiding singular matrices when plotting distributions
                        self.results.loc[(full_name,i),"value"] = (max(min(float(val) if not np.isinf(val) or np.isnan(val) else 1e20,1e20),-1e20))+np.random.random()*1e-20
                        self.results.loc[(full_name,i),"time"] = float(elapsed_time)
                        self.results.loc[(full_name,i),"reads"] = int(reads)+np.random.random()*1e-20
                        error,dist,found = self.compute_error()
                        self.results.loc[(full_name,i),"error"] =max(min(float(error),1e20),-1e20)+np.random.random()*1e-20
                        self.results.loc[(full_name,i),"dist"] = max(min(float(dist),1e20),-1e20)+np.random.random()*1e-20
                        self.results.loc[(full_name,i),"found"] = int(found)
                        self.func_pb.value = j+1
                        if self.val_every.value == 'Function' and np.mod(self.num_every.value,j) == 0:
                            self.save_data()
                    if self.val_every.value == 'Round':
                        self.save_data()
                    self.run_pb.value = i+1
                    subset = self.results['full_id'] == full_name
                    self._html_display_value = self.results[subset].describe().to_html()
                    self.update_html(name=full_name)
                    #time.sleep(0.15)
                    
        
            if self.val_every.value == 'Algorithm' and np.mod(self.num_every.value,k) == 0:
                self.save_data()
        self.save_data()
        self.update_html()
        clear_output(True) 
        
    
    def save_data(self,path='data/'):
        """Saves data to pickle"""
        self.results.dropna().to_pickle('{}{}.pck'.format(path,self.save_name.value))
        
    def update_html(self,_=None, name=''):
        """Updates the summary results of the optimization"""
        self.plot.value =  name+'<div style="width:100%;height:100%">'+self._html_display_value+'</div>'
        
    def run_one(self,func,name):
        """Solve one function with one algorithm"""
        self.optimizer.pot_func = func
        self.optimizer.reset()
        self._run_start = time.time()
        self.optimizer.run(name=name,end_callback=self.end_condition)
        x,val,reads = self.optimizer.best[0],self.optimizer.best[1],self.optimizer.pot_func.n_reads
        elapsed_time = float(time.time()-self._run_start)
        return  x,val,reads,elapsed_time
```
